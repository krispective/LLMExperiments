```python
import requests
from bs4 import BeautifulSoup

def scrape_straighttalk_phones(url="https://www.straighttalk.com/all-phones"):
    headers = {
        "User-Agent": "Mozilla/5.0"
    }

    response = requests.get(url, headers=headers)
    response.raise_for_status()
    soup = BeautifulSoup(response.text, 'lxml')

    phones_data = []

    # Each phone appears in a card with class like "product-tile" or similar
    phone_cards = soup.find_all('div', class_='product-tile')  # May need to adjust class

    for card in phone_cards:
        name = card.find('a', class_='product-name')
        price = card.find('span', class_='price') or card.find('span', class_='sr-only')
        description = card.find('div', class_='product-description')

        phone_info = {
            "name": name.get_text(strip=True) if name else "N/A",
            "price": price.get_text(strip=True) if price else "N/A",
            "description": description.get_text(strip=True) if description else "N/A"
        }
        phones_data.append(phone_info)

    return phones_data

# Format data for RAG
def format_for_rag(phones_data):
    summaries = []
    for phone in phones_data:
        summary = f"Phone: {phone['name']}\nPrice: {phone['price']}\nDescription: {phone['description']}\n"
        summaries.append(summary)
    return "\n---\n".join(summaries)
```

```python
scraped_data = scrape_straighttalk_phones()
formatted_summary = format_for_rag(scraped_data)
print(formatted_summary)
```
